```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df_movies = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
df_credits = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv")
```

```python
tfidf = TfidfVectorizer(stop_words="english")
df_movies['overview'] = df_movies['overview'].fillna("")
tfidf_matrix = tfidf.fit_transform(df_movies['overview'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df_movies.index, index=df_movies['original_title']).drop_duplicates()
indices
```

```python
indices['John Carter']
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    # Access values directly or use a different looping mechanism:
    sim_scores = [(i, cosine_sim[idx][i]) for i in range(len(cosine_sim[idx]))]
    # Sort using the second element (value) of each tuple:
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Select top 10 indices (assuming `sim_scores` is now a list of tuples):
    sim_index = [i for i, _ in sim_scores[:10]]  # Extract only the first element (index)
    # Ensure valid integer indices within df_movies range:
    sim_index = [i for i in sim_index if 0 <= i < len(df_movies["original_title"])]
    # Print recommended titles:
    print(df_movies["original_title"].iloc[sim_index])
```

```python
get_recommendations('John Carter')
```
